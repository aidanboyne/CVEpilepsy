### Video annotation creator

Run this first to make a nice file for the clip/annotaion file creation notebook from the video notes

```python
import os

annotation_file = "annotations/video_annotations.txt"
video_annot = []
video_notes = "annotations/usuable_data.txt" 

# Check if the annotation file exists and delete it
if os.path.exists(annotation_file):
    os.remove(annotation_file)

with open(video_notes, "r") as file:
    patient_id = None
    for line in file:
        line = line.strip()

        # Check for patient ID line
        if line.startswith("Patient"):
            patient_id = line.split("(")[1].strip("):")
            print(f"Processing patient: {patient_id}")
            continue
        
        # Split the line at the comment delimiter
        video_info, *_ = line.split("#", 1)

        # Parse the line for video information
        try:
            parts = video_info.strip("() ").split(",")
            video_path, start_time, end_time, video_id, unit = parts
            start_time, end_time = int(start_time), int(end_time)

            # Create the annotation tuple
            annotation_tuple = (
                os.path.join(video_path.strip("'")),
                video_id.strip("'"),
                patient_id,
                [start_time, end_time]
            )

            # Add the annotation tuple to the list
            video_annot.append(annotation_tuple)
            print(f"Added annotation: {annotation_tuple}")
        except ValueError:
            print(f"Skipping line: {line}")

# Write the annotations to the annotation file 
with open(annotation_file, "w") as file:
    for annotation in video_annot:
        file.write(str(annotation) + "\n")
        print(f"Wrote annotation: {annotation}")
```
